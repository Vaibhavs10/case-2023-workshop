```python
from transformers import pipeline
from datasets import load_dataset
```

```python
pipe = pipeline("automatic-speech-recognition",
                "openai/whisper-tiny.en")
```

```python
ls = load_dataset("hf-internal-testing/librispeech_asr_dummy",
                  "clean",
                  split="validation")
```

```python
ls[0]
```

```python
import IPython.display as ipd

ipd.Audio(data=ls[0]["audio"]["array"], autoplay=False, rate=ls[0]["audio"]["sampling_rate"])
```

```python
pipe(ls[0]["audio"])
```

```python
pipe = pipeline("text-to-speech",
                "suno/bark-small")
```

```python
output = pipe("How is it going y'all?")
```

```python
output
```

```python
ipd.Audio(data=output["audio"], autoplay=False, rate=output["sampling_rate"])
```
